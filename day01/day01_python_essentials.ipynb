## 1. Booleans & Conditionals

A **boolean** is a value that is either `True` or `False`. Booleans are the foundation of decision-making in code.

````python
is_sunny = True
is_raining = False

print(is_sunny)
print(type(is_sunny))
````

### Comparison operators

You create booleans by **comparing** values. Here are the comparison operators:

| Operator | Meaning |
|---|---|
| `==` | equal to |
| `!=` | not equal to |
| `<` | less than |
| `>` | greater than |
| `<=` | less than or equal to |
| `>=` | greater than or equal to |

**Important:** `=` *assigns* a value. `==` *compares* two values.

````python
temperature = 72

print(temperature > 80)
print(temperature == 72)
print(temperature != 72)
print(temperature <= 100)
````

### `if`, `elif`, `else`

Booleans let us write code that makes **decisions**. The pattern is:

```python
if condition:
    # do this
elif other_condition:
    # do this instead
else:
    # do this if nothing above was True
```

Python uses **indentation** (4 spaces) to know which code belongs inside each block.

````python
temperature = 85

if temperature > 90:
    print("It's really hot!")
elif temperature > 70:
    print("It's a nice day.")
else:
    print("You might want a jacket.")
````

Try changing the value of `temperature` above and re-running the cell. Which branch runs?

Remember from the previous notebook: when you change a variable and re-run its cell, the new value is now in Python's memory. But any cell that *already used* the old value won't update automatically — you need to re-run those cells too.

### Combining conditions with `and`, `or`, `not`

You can combine multiple conditions:

- `and` — both must be True
- `or` — at least one must be True
- `not` — flips True to False (and vice versa)

````python
temperature = 75
is_sunny = True

if temperature > 60 and is_sunny:
    print("Great day for a walk!")

if temperature > 100 or is_sunny:
    print("You'll want sunscreen.")

if not is_raining:
    print("No umbrella needed.")
````

**Why this matters for data science:** When we get to pandas, we'll filter data with conditions like `df[df["age"] > 30]`. The `>` produces a column of booleans, and pandas keeps only the `True` rows. Same idea, bigger scale!

## 2. Imports

Python comes with a lot of built-in functions (`print`, `len`, `sum`, etc.), but most of its power comes from **libraries** — collections of code that other people have written.

To use a library, you **import** it.

````python
import math

print(math.sqrt(144))
print(math.pi)
````

`math` is a library that comes with Python. After importing it, we access its functions and values with a dot: `math.sqrt()`, `math.pi`.

### Aliasing with `as`

Some library names are long, so we give them a shorter **alias**. This is just a nickname.

````python
import math as m

# Now we can use m instead of math
print(m.sqrt(144))
````

You'll see this pattern constantly in data science:

```python
import pandas as pd        # pd is the standard alias for pandas
import matplotlib.pyplot as plt  # plt is the standard alias for matplotlib
import numpy as np         # np is the standard alias for numpy
```

These aliases are conventions — everyone uses them, so you'll recognize them in tutorials and documentation.

### Importing specific things with `from`

If you only need one or two things from a library, you can import them directly.

````python
from math import sqrt, pi

# Now we can use sqrt and pi directly, without the math. prefix
print(sqrt(144))
print(pi)
````

> **Common beginner mistake:** using a library before running the `import` cell. If you see a `NameError` for something like `pd`, `plt`, or `np` later in this course, scroll up and make sure you ran the import cell first. Imports are just regular code cells — they have to be **executed** before anything that uses them.

### Built-in vs. external libraries

- **Built-in** libraries (like `math`, `string`, `random`) come with Python — just `import` them.
- **External** libraries (like `pandas`, `matplotlib`) need to be **installed** first, usually with `pip install pandas` in your terminal. In this workshop, they're already installed for you!

## 3. String Methods & f-strings

In the previous notebook we saw that strings are sequences of characters. Python gives us many built-in **methods** for working with strings.

A **method** is a function that belongs to an object — you call it with a dot: `my_string.upper()`.

````python
message = "  Hello, World!  "

print(message.upper())
print(message.lower())
print(message.strip())       # Remove leading/trailing whitespace
print(message.replace("World", "Python"))
````

````python
# .split() breaks a string into a list of words
sentence = "data science is awesome"
words = sentence.split()
print(words)
print(len(words))
````

### f-strings: formatted string literals

An **f-string** lets you embed variables directly inside a string. Just put `f` before the opening quote and use `{}` around your variables.

````python
name = "Ada"
age = 28
city = "Ann Arbor"

# Without f-strings (clunky)
print("My name is " + name + " and I am " + str(age) + " years old.")

# With f-strings (much cleaner!)
print(f"My name is {name} and I am {age} years old.")
print(f"{name} lives in {city}.")
````

You can put any expression inside the `{}`:

````python
price = 19.99
quantity = 3

print(f"Total: ${price * quantity:.2f}")  # .2f means 2 decimal places
````

### Type conversion

Sometimes you need to convert between types. Python gives you `str()`, `int()`, and `float()` for this.

````python
x = "42"
print(type(x))        # It's a string!

x_as_int = int(x)
print(type(x_as_int)) # Now it's an integer
print(x_as_int + 8)   # And we can do math with it
````

## 4. List & Dictionary Operations

In the previous notebook we created lists and dictionaries. Now let's learn how to **modify** and **search** them.

### Building up a list

Use `.append()` to add one item at a time.

````python
fruits = ["apple", "banana"]
print(f"Before: {fruits}")

fruits.append("cherry")
fruits.append("date")
print(f"After: {fruits}")
````

### Slicing lists

You can grab a **slice** (a sub-section) of a list using `[start:stop]`. The `start` index is included, but `stop` is **not** included.

````python
numbers = [10, 20, 30, 40, 50, 60]

print(numbers[1:4])   # Items at index 1, 2, 3
print(numbers[:3])    # First 3 items (start defaults to 0)
print(numbers[3:])    # Everything from index 3 onward
print(numbers[-1])    # Last item
print(numbers[-2:])   # Last 2 items
````

### Checking membership with `in`

Use `in` to check whether something is in a list (or a string, or a dictionary).

````python
print("banana" in fruits)
print("mango" in fruits)

# This is great for conditionals!
if "cherry" in fruits:
    print("We have cherries!")
````

### List comprehensions (a preview)

A **list comprehension** is a compact way to create a new list by transforming each item in an existing one. You'll see these in data science code, so let's get familiar.

````python
numbers = [1, 2, 3, 4, 5]

# The long way
squares = []
for n in numbers:
    squares.append(n ** 2)
print(squares)

# The list comprehension way (same result, one line)
squares = [n ** 2 for n in numbers]
print(squares)
````

You can even add a condition:

````python
# Only keep the even numbers
evens = [n for n in numbers if n % 2 == 0]
print(evens)
````

### Dictionary operations

Dictionaries store **key-value pairs**. Let's see how to inspect and modify them.

````python
populations = {"Ann Arbor": 123851, "Detroit": 639111, "Lansing": 112644}

# Look up a value
print(populations["Detroit"])

# Add a new entry
populations["Grand Rapids"] = 198917
print(populations)
````

````python
# Get all the keys, values, or both
print(list(populations.keys()))
print(list(populations.values()))

# .items() gives you key-value pairs — great for looping
for city, pop in populations.items():
    print(f"{city}: {pop:,} people")  # :, adds commas to numbers
````

````python
# Check if a key exists
print("Detroit" in populations)
print("Chicago" in populations)
````

## 5. Defining Functions

So far we've used built-in functions like `print()`, `len()`, and `sum()`. You can also define your **own** functions to package up a piece of logic and reuse it.

```python
def function_name(parameter1, parameter2):
    # do something
    return result
```

- `def` starts the definition
- The **parameters** are inputs the function accepts
- `return` sends a value back to the caller

````python
def fahrenheit_to_celsius(temp_f):
    """Convert a temperature from Fahrenheit to Celsius."""
    temp_c = (temp_f - 32) * 5/9
    return temp_c

# Now we can use it like any other function
print(fahrenheit_to_celsius(32))
print(fahrenheit_to_celsius(72))
print(fahrenheit_to_celsius(212))
````

Once defined, you can call your function as many times as you like. The `"""..."""` on the second line is a **docstring** — a short description of what the function does.

Functions can take multiple parameters:

````python
def greet(name, greeting="Hello"):
    """Return a greeting for the given name."""
    return f"{greeting}, {name}!"

print(greet("Ada"))
print(greet("Ada", "Welcome"))
````
