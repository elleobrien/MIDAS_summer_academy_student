### Notebook gotcha: Out-of-order execution

Notebooks let you run cells in **any order**. This is powerful, but it can cause confusing bugs. Try this experiment:

1. Run the three code cells below **in order** — you should see `Total is 30`.
2. Now go back and run **only the second** code cell (`price = 99`).
3. Then re-run the third cell. What happened?

```python
price = 10
```

```python
price = 99  # Imagine you ran this cell while experimenting
```

```python
total = price * 3
print(f"Total is {total} — is that what you expected?")
```

The value of `total` depends on which `price` cell you ran **most recently**, not which one appears first in the notebook. This is called **stale state** — a variable holds a value you didn't expect because cells were run out of order.

When you get a surprising result, ask yourself: *"Did I run these cells in order from top to bottom?"*

A quick fix: **Kernel > Restart & Run All** (or **Runtime > Restart and run all** in Colab) re-runs everything top-to-bottom, clearing out any stale state.

```python
# Python's string module has useful constants like the alphabet
import string
alphabet = list(string.ascii_lowercase)
print("The 14th letter is:", alphabet[13])
```
